```python
##Subplots
plt.figure(figsize=(10,6))
def laplace(x, lam):
	return (1/(2*lam)) * np.e**(-1*(np.abs(x)/lam))

def normal(x, mean, sigma):
	return (np.e**(-(x-mean)**2/(2*sigma**2)))/(sigma*np.sqrt(2*np.pi))
    
x = np.linspace(-4,4,10000)
y1 = [laplace(i, 0.5) for i in x]
#now create a subplot which represents the top plot of a grid
# with 2 rows and 2 columns. 
plt.subplot(2,2,1)
#set the y-limits of the current axes
plt.ylim(0,1.1)
plt.plot(x, y1, linewidth=1, color='r',label='lambda=0.5')
# Place a legend on the axes
plt.legend()

y2 = [laplace(i, 1) for i in x]
plt.subplot(2,2,2)
plt.ylim(0,1.1)
plt.plot(x, y2, linewidth=1, color='r', label='lambda=1')
plt.legend()

y3 = [normal(i, 0, 1) for i in x]
plt.subplot(2,2,3)
plt.ylim(0,0.45)
a=plt.plot(x, y3, linewidth=1, color='r', label='u=0,sigma=1')
plt.legend()

y4 = [normal(i, 0, 2) for i in x]
plt.subplot(2,2,4)
plt.ylim(0,0.45)
plt.plot(x, y4, linewidth=1, color='r',label='u=0,sigma=2')
plt.legend()
```

#### 10.2 Sinusoidal data sets with noise

```python
from matplotlib import pyplot as pt
import numpy as np
import random

def no_noise(X):
    return np.sin(X)

X = np.linspace(-3, 3, 50, endpoint=True)
Y = np.sin(X)


mu = 0
sigma = 0.08
for i in range(X.size):

    Y[i] += random.gauss(mu, sigma)
    #Draw random samples from a normal (Gaussian) distribution.


pt.scatter(X,Y)
pt.plot(X,no_noise(X))
pt.title("Sinusoidal data sets with noise")
pt.show()
```

### 11 Using gradient descent to solve optimization problem

#### ð‘”(ð‘¤)=sinâ¡(3ð‘¤)+0.3ð‘¤^2

```python
#5-1 
#ð‘”(ð‘¤)=sinâ¡(3ð‘¤)+0.3ð‘¤^2

import matplotlib.pyplot as plt
import numpy as np
from numpy.ma import exp,sin,cos



def func_1d(x):
    #objective function
    return sin(3*x)+0.3*(x**2)


def grad_1d(x):
    #gradient of the function
    return cos(3*x)*3+0.6*x


def gradient_descent_1d(grad, cur_x, learning_rate, precision, max_iters, history_x):
    """
    :param grad: gradient
    :param prev_x: x value from previous iteration {initalized with cur_x}
    :param cur_x: current x value
    :param learning_rate: learning rate
    :param precision: convergence precision
    :param max_iters: maximum number of iterations
    :param history_x: track the change of x
    :return: x changes until convergence
    """
    print(f"The initial x value is: {cur_x}")
    
    for i in range(max_iters):
        grad_cur = grad(cur_x)
        prev_x = cur_x
        cur_x = cur_x - grad_cur * learning_rate
        history_x.append(cur_x)
        if abs(func_1d(prev_x)-func_1d(cur_x)) < precision:
            #convergence
            break  
            
        print(f"At iteration {i}, x is: {cur_x}")

    print(f"The local minimum exists at: ({cur_x},{func_1d(cur_x)})")
    
    return history_x
plot_x=np.linspace(-1,0)   
plot_y=func_1d(plot_x)
a = gradient_descent_1d(grad_1d, cur_x=-1, learning_rate=0.1, precision=1e-8, max_iters=10000,history_x=[-1])
plt.plot(plot_x,plot_y)    # plot the objective function
plt.plot(np.array(a),func_1d(np.array(a)),color='r',marker='*')  #plot the convergence track 
plt.grid()
plt.show()

```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy.ma import exp,sin,cos

sum_eta=[]
result=[]
for i in range(1,20,1):
    #when the learning rate change from 0.01 to 0.2
    x=-1                           
    eta=i*0.01
    sum_eta.append(eta)
    epsilon=1e-8                    
    num=0
    min=0
    while True:
        gradient=cos(3*x)*3+0.6*x            
        last_x=x
        x=x-eta*gradient
        num=num+1
        if (abs(func_1d(last_x)-func_1d(x)) <epsilon):       
            min=x
            break
     
    result.append(num) #track the number of iterations
 
plt.scatter(sum_eta,result,c='r')
plt.plot(sum_eta,result,c='r')
plt.title("relation")
plt.xlabel("learning rate")
plt.ylabel("iteration times")
plt.grid()
plt.show()
print(result)
min = np.min(result)
index = result.index(min)
learningRate = sum_eta[index]
print(f"The learning rate with the least # of iterations was: {learningRate} with an iteration count of: {min}")
```

#### ð‘”(ð‘¤_1,ð‘¤_2 )=ð‘¤_1^2+ð‘¤_2^2+2

```python
#ð‘”(ð‘¤_1,ð‘¤_2 )=ð‘¤_1^2+ð‘¤_2^2+2
import math
import numpy as np


def func_2d(x):
    #objective function
    return x[0] ** 2 + x[1] ** 2+2


def grad_2d(x):
    #gradient of the function
    deriv0 = 2 * x[0]
    deriv1 = 2 * x[1] 
    return np.array([deriv0, deriv1])


def gradient_descent_2d(grad, cur_x, learning_rate, precision, max_iters,history_x):
    """
    :param grad: gradient
    :param prev_x: x value from previous iteration {initalized with cur_x}
    :param cur_x: current x value
    :param learning_rate: learning rate
    :param precision: convergence precision
    :param max_iters: maximum number of iterations
    :param history_x: track the change of x
    :return: x changes until convergence
    """
    print(f"The initial x value is: {cur_x}")
    for i in range(max_iters):
        grad_cur = grad(cur_x)
        history_x.append(cur_x) #track the change of x
        last_x = cur_x
        cur_x = cur_x - grad_cur * learning_rate
        if abs(func_2d(last_x)-func_2d(cur_x)) < precision:
            break  # when the difference close to 0, we assume the convergence
        print(f"At iteration {i}, x is: {cur_x}")

    #print("[x1,x2] =", cur_x)
    #print("local minimum ",func_2d(cur_x))
    print(f"The local minimum exists at: ({cur_x},{func_2d(cur_x)})")
    return history_x

a = gradient_descent_2d(grad_2d, cur_x=np.array([1,1]), learning_rate = 0.3, precision=0.000001, max_iters=10000,history_x=[])


from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig=plt.figure(figsize = (9,7))
ax1 = plt.axes(projection='3d')
xd = np.linspace(-1,1)   
yd = np.linspace(-1,1)   
zd = xd ** 2 + yd ** 2+2

x0 = np.array(a)[:,0]
x1 = np.array(a)[:,1]
z = x0 ** 2 + x1 ** 2+2

ax1.plot3D(xd,yd,zd) #plot the objective function
ax1.plot3D(x0,x1,z,color='r',marker='*',label = 'learning rate = 0.3') #plot the convergence track
ax1.legend()
    
plt.show()
```
