```python
import numpy as np

x = np.array([-2.4, 1.5, 0.0, 3.2])
y = np.array([1.3, 4.2, 3.0, -2.5])

norm1 = np.linalg.norm(x, 1)
norm2 = np.linalg.norm(x, 2)
norm3 = np.linalg.norm(x, 3)
norm_inf = np.linalg.norm(x, np.inf)

norms = [norm1, norm2, norm3, norm_inf]
norms_sorted = sorted(norms, reverse=True)

dist1 = np.linalg.norm(x - y, 1)
dist2 = np.linalg.norm(x - y, 2)
dist_inf = np.linalg.norm(x - y, np.inf)

distances = [dist1, dist2, dist_inf]
distances_sorted = sorted(distances)

print("Part (a) Norms of vector x in descending order:")
print(norms_sorted)

print("\nPart (b) Distances between x and y in ascending order:")
print(distances_sorted)
```

### Problem 3:

```python
import numpy as np
from numpy import linalg as LA

B = np.array([[2, 2, 4],
              [1, 4, 7],
              [2, 5, 4]])

eigenvalues, eigenvectors = np.linalg.eig(B)

Q = eigenvectors
Lamda = np.diag(eigenvalues)
Q_inv = np.linalg.inv(Q)

B_calculated = Q @ Lamda @ Q_inv
res_of_verification = np.allclose(B, B_calculated)

print("Eigenvalues of matrix B:")
print(eigenvalues)

print("\nEigenvectors of matrix B from columns of Q:")
print(eigenvectors)

print("\nMatrix of eigenvalues Λ:")
print(Lamda)

print("\nInverse of matrix Q:")
print(Q_inv)

print("\nCalculated matrix B:")
print(B_calculated)

print("\nVerify that B = Q Λ Q_inverse:")
print(res_of_verification)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return pow(x, 2) + 2*pow(y, 2) - 2*x*y

def gradient(x, y):
    df_dx = 2*x - 2*y
    df_dy = 4*y - 2*x
    return np.array([df_dx, df_dy])

def gradient_descent(learning_rate, initial_point, max_iters=100, tol=1e-6):
    x, y = initial_point
    record = [(x, y)]
    for _ in range(max_iters):
        grad = gradient(x, y)
        new_x = x - learning_rate * grad[0]
        new_y = y - learning_rate * grad[1]
        
        if np.linalg.norm(np.array([new_x - x, new_y - y])) < tol:
            break
            
        x, y = new_x, new_y
        record.append((x, y))
        
    return np.array(record)

initial_point = np.array([1.0, 1.0])
learning_rates = [0.1, 0.01, 0.001]

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

for lr in learning_rates:
    record = gradient_descent(lr, initial_point)
    iterations = np.arange(len(record))

    axs[0].plot(iterations, record[:, 0], label=f'Learning rate: {lr}')
    axs[0].set_title('Convergence of x')
    axs[0].set_xlabel('Number of interations')
    axs[0].set_ylabel('x')

    axs[1].plot(iterations, record[:, 1], label=f'Learning rate: {lr}')
    axs[1].set_title('Convergence of y')
    axs[1].set_xlabel('Number of interations')
    axs[1].set_ylabel('y')

axs[0].legend()
axs[1].legend()
axs[0].grid(True)
axs[1].grid(True)
plt.tight_layout()
```
